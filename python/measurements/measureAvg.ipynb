```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Measures_Human.xlsx'
df = pd.read_excel(file_path)


avgDfHuman = df[['Number', 'extrovertism avg', 'honesty avg', 'neuroticism avg', 'rigidity avg']].copy()

avgDfHuman['extrovertism avg'] = pd.to_numeric(avgDfHuman['extrovertism avg'], errors='coerce')
avgDfHuman['honesty avg'] = pd.to_numeric(avgDfHuman['honesty avg'], errors='coerce')
avgDfHuman['neuroticism avg'] = pd.to_numeric(avgDfHuman['neuroticism avg'], errors='coerce')
avgDfHuman['rigidity avg'] = pd.to_numeric(avgDfHuman['rigidity avg'], errors='coerce')

avgDfHuman.rename(columns={
    'extrovertism avg': 'extrovertism_human',
    'honesty avg': 'honesty_human',
    'neuroticism avg': 'neuroticism_human',
    'rigidity avg': 'rigidity_human'
}, inplace=True)

plt.figure(figsize=(10, 6))
avgDfHuman[['extrovertism_human', 'honesty_human', 'neuroticism_human', 'rigidity_human']].boxplot()
plt.title('Average for personality types of Humans')
plt.ylabel('Average')
plt.xlabel('Personality types')
plt.grid(True)
plt.show()
```

```python
file_path = 'Measures_LLM.xlsx'
df = pd.read_excel(file_path)

avgDfLLM = df[['LLM', 'extrovertism avg', 'honesty avg', 'neuroticism avg', 'rigidity avg']].copy()

avgDfLLM['extrovertism avg'] = pd.to_numeric(avgDfLLM['extrovertism avg'], errors='coerce')
avgDfLLM['honesty avg'] = pd.to_numeric(avgDfLLM['honesty avg'], errors='coerce')
avgDfLLM['neuroticism avg'] = pd.to_numeric(avgDfLLM['neuroticism avg'], errors='coerce')
avgDfLLM['rigidity avg'] = pd.to_numeric(avgDfLLM['rigidity avg'], errors='coerce')

avgDfLLM.rename(columns={
    'extrovertism avg': 'extrovertism_LLM',
    'honesty avg': 'honesty_LLM',
    'neuroticism avg': 'neuroticism_LLM',
    'rigidity avg': 'rigidity_LLM'
}, inplace=True)

plt.figure(figsize=(10, 6))
avgDfLLM[['extrovertism_LLM', 'honesty_LLM', 'neuroticism_LLM', 'rigidity_LLM']].boxplot()
plt.title('Average for personality types of LLMs')
plt.ylabel('Average')
plt.xlabel('Personality types')
plt.grid(True)
plt.show()
```

```python
combined_df = pd.concat([avgDfHuman, avgDfLLM], axis=1)

plt.figure(figsize=(10, 6))
combined_df.boxplot(column=[
    'extrovertism_human', 'extrovertism_LLM',
    'honesty_human', 'honesty_LLM',
    'neuroticism_human', 'neuroticism_LLM',
        'rigidity_human', 'rigidity_LLM'
])
plt.title('Average for personality types of Humans and LLMs')
plt.ylabel('Average')
plt.xlabel('Personality Types')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
